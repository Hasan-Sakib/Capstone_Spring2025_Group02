```python
# =========================
# Imports
# =========================
import os, time, random
import numpy as np
from glob import glob

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

import torchvision.models as models
import torchvision.transforms as T

from torch.utils.data import DataLoader, Dataset, WeightedRandomSampler

from PIL import Image
from tqdm import tqdm

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.manifold import TSNE
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
import joblib   # <-- added for saving/loading Random Forest

# Reproducibility
seed = 42
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)
torch.cuda.manual_seed_all(seed)

# =========================
# Paths & Hyperparams
# =========================
root_dir = "/content/drive/MyDrive/Colab Notebooks/malimg_paper_dataset_imgs"

batch_size = 128
pretrain_epochs = 100
learning_rate = 1e-3
temperature  = 0.5
projection_dim = 128
test_ratio = 0.2
patience = 5
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# =========================
# Dataset helpers
# =========================
def list_images_and_labels(root_dir):
    class_names = sorted([d for d in os.listdir(root_dir)
                          if os.path.isdir(os.path.join(root_dir, d))])
    class_to_idx = {c:i for i,c in enumerate(class_names)}
    img_paths, labels = [], []
    for cls in class_names:
        cls_path = os.path.join(root_dir, cls)
        for f in os.listdir(cls_path):
            p = os.path.join(cls_path, f)
            if os.path.isfile(p):
                img_paths.append(p)
                labels.append(class_to_idx[cls])
    return img_paths, labels, class_names

class MalimgDatasetFromList(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img = Image.open(self.image_paths[idx]).convert("RGB")
        label = self.labels[idx]
        if self.transform:
            img = self.transform(img)
        return img, label

class ContrastiveDataset(Dataset):
    def __init__(self, base_dataset, augment):
        self.base = base_dataset
        self.augment = augment

    def __len__(self):
        return len(self.base)

    def __getitem__(self, idx):
        img, _ = self.base[idx]
        if hasattr(self.base, "image_paths"):
            pil = Image.open(self.base.image_paths[idx]).convert("RGB")
        else:
            pil = img
        x1 = self.augment(pil)
        x2 = self.augment(pil)
        return x1, x2

# =========================
# Model (SimCLR wrapper)
# =========================
class SimCLR(nn.Module):
    def __init__(self, base_model, projection_dim=128):
        super().__init__()
        self.encoder = base_model
        self.projection = nn.Sequential(
            nn.Linear(1000, 512),
            nn.ReLU(inplace=True),
            nn.Linear(512, projection_dim)
        )

    def forward(self, x):
        h = self.encoder(x)
        z = self.projection(h)
        return h, z

def nt_xent_loss(z1, z2, temperature=0.5):
    N = z1.size(0)
    z1 = F.normalize(z1, dim=1)
    z2 = F.normalize(z2, dim=1)
    z  = torch.cat([z1, z2], dim=0)

    sim = torch.matmul(z, z.T) / temperature
    mask = torch.eye(2*N, dtype=torch.bool, device=z.device)
    sim = sim.masked_fill(mask, -9e15)

    pos = torch.cat([torch.diag(sim, N), torch.diag(sim, -N)], dim=0)
    neg = sim

    logits = torch.cat([pos.unsqueeze(1), neg], dim=1)
    labels = torch.zeros(2*N, dtype=torch.long, device=z.device)
    loss = F.cross_entropy(logits, labels)
    return loss

# =========================
# Data prep + split
# =========================
all_paths, all_labels, class_names = list_images_and_labels(root_dir)
num_classes = len(class_names)

train_paths, test_paths, train_labels, test_labels = train_test_split(
    all_paths, all_labels,
    test_size=test_ratio,
    stratify=all_labels,
    random_state=seed
)

train_dataset_pil = MalimgDatasetFromList(train_paths, train_labels, transform=None)

simclr_augment = T.Compose([
    T.RandomResizedCrop(224),
    T.RandomHorizontalFlip(),
    T.RandomApply([T.ColorJitter(0.4, 0.4, 0.4, 0.1)], p=0.8),
    T.RandomGrayscale(p=0.2),
    T.ToTensor()
])

contrastive_dataset = ContrastiveDataset(train_dataset_pil, simclr_augment)

class_counts = np.bincount(train_labels, minlength=num_classes).astype(np.float32)
class_counts[class_counts == 0] = 1.0
weights_per_class = 1.0 / class_counts
sample_weights = np.array([weights_per_class[y] for y in train_labels], dtype=np.float32)
train_sampler = WeightedRandomSampler(
    weights=torch.as_tensor(sample_weights),
    num_samples=len(sample_weights),
    replacement=True
)

contrastive_loader = DataLoader(
    contrastive_dataset,
    batch_size=batch_size,
    sampler=train_sampler,
    num_workers=2,
    pin_memory=True
)

train_transform_cls = T.Compose([
    T.Resize((224, 224)),
    T.ToTensor()
])
test_transform_cls = T.Compose([
    T.Resize((224, 224)),
    T.ToTensor()
])

train_dataset_cls = MalimgDatasetFromList(train_paths, train_labels, transform=train_transform_cls)
test_dataset_cls  = MalimgDatasetFromList(test_paths,  test_labels,  transform=test_transform_cls)

train_loader_cls = DataLoader(
    train_dataset_cls,
    batch_size=batch_size,
    sampler=train_sampler,
    num_workers=2,
    pin_memory=True
)
test_loader = DataLoader(
    test_dataset_cls,
    batch_size=batch_size,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)

# =========================
# Build models
# =========================
base_model = models.resnet50(pretrained=True)
simclr_model = SimCLR(base_model, projection_dim=projection_dim).to(device)

optimizer = optim.Adam(simclr_model.parameters(), lr=learning_rate)

# =========================
# Pretraining (SimCLR) + timing
# =========================
pretrain_start = time.time()
best_train_loss = float("inf")
no_improve = 0

for epoch in range(pretrain_epochs):
    simclr_model.train()
    running_loss = 0.0
    for (x1, x2) in tqdm(contrastive_loader, desc=f"SimCLR Pretrain {epoch+1}/{pretrain_epochs}"):
        x1, x2 = x1.to(device, non_blocking=True), x2.to(device, non_blocking=True)

        _, z1 = simclr_model(x1)
        _, z2 = simclr_model(x2)
        loss = nt_xent_loss(z1, z2, temperature=temperature)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    epoch_loss = running_loss / max(1, len(contrastive_loader))
    print(f"[Pretrain] Epoch {epoch+1}/{pretrain_epochs} | Loss: {epoch_loss:.4f}")

    if epoch_loss + 1e-6 < best_train_loss:
        best_train_loss = epoch_loss
        no_improve = 0
    else:
        no_improve += 1
        if no_improve >= patience:
            print(f"Early stopping SimCLR pretraining at epoch {epoch+1} (no improvement for {patience} epochs).")
            break

if device.type == "cuda":
    torch.cuda.synchronize()
pretrain_time = time.time() - pretrain_start
print(f"Pretraining time: {pretrain_time/60:.2f} min")

# =========================
# Feature extraction for Random Forest
# =========================
def extract_features(dataloader, encoder_model):
    feats, labels = [], []
    encoder_model.eval()
    with torch.no_grad():
        for imgs, lbls in tqdm(dataloader, desc="Extracting features"):
            imgs = imgs.to(device)
            h, _ = encoder_model(imgs)   # encoder output (1000-dim)
            feats.append(h.cpu().numpy())
            labels.extend(lbls.numpy())
    return np.vstack(feats), np.array(labels)

print("Extracting train features...")
X_train, y_train = extract_features(train_loader_cls, simclr_model)
print("Extracting test features...")
X_test, y_test = extract_features(test_loader, simclr_model)

# =========================
# Train Random Forest Classifier
# =========================
print("Training Random Forest...")
rf_clf = RandomForestClassifier(
    n_estimators=300,
    max_depth=None,
    class_weight="balanced",
    n_jobs=-1,
    random_state=seed
)
rf_clf.fit(X_train, y_train)

# =========================
# Train Accuracy
# =========================
y_train_pred = rf_clf.predict(X_train)
train_acc = accuracy_score(y_train, y_train_pred) * 100.0
print(f"Train Accuracy (Random Forest): {train_acc:.2f}%")

# =========================
# Test Evaluation with Inference Timing
# =========================
print("Evaluating Random Forest...")
start_time = time.time()
y_pred = rf_clf.predict(X_test)
end_time = time.time()

test_acc = accuracy_score(y_test, y_pred) * 100.0
avg_infer_ms = ((end_time - start_time) / max(1, len(y_test))) * 1000.0

print(f"Test Accuracy: {test_acc:.2f}%")
print(f"Average Inference Time (per image): {avg_infer_ms:.3f} ms")

print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=class_names, digits=4))

cm = confusion_matrix(y_test, y_pred, labels=list(range(num_classes)))
plt.figure(figsize=(10, 8))
plt.imshow(cm, interpolation='nearest')
plt.title("Confusion Matrix (Random Forest)")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.colorbar()
tick_marks = np.arange(num_classes)
plt.xticks(tick_marks, class_names, rotation=90)
plt.yticks(tick_marks, class_names)
plt.tight_layout()
plt.show()

# =========================
# t-SNE visualization
# =========================
tsne = TSNE(n_components=2, perplexity=30, learning_rate=200, random_state=seed, n_iter=1000)
features_2d = tsne.fit_transform(X_test)

plt.figure(figsize=(10, 8))
plt.scatter(features_2d[:, 0], features_2d[:, 1], c=y_test, s=10, alpha=0.7)
plt.title("t-SNE of Encoder Features (Test Set, Random Forest)")
plt.tight_layout()
plt.show()

# =========================
# Save & Load Random Forest
# =========================
save_path = "/content/drive/MyDrive/Colab Notebooks/SimCLR/random_forest_clf.pkl"
joblib.dump(rf_clf, save_path)
print(f"Random Forest model saved to {save_path}")

# Example load (uncomment when needed):
# rf_clf = joblib.load(save_path)
# print("Random Forest model loaded successfully!")

print("Done. Random Forest training + evaluation complete.")
```
