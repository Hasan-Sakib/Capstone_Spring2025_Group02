```python

# =========================
# Imports
# =========================
import os, time, random
import numpy as np
from glob import glob

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

import torchvision.models as models
import torchvision.transforms as T

from torch.utils.data import DataLoader, Dataset, WeightedRandomSampler

from PIL import Image
from tqdm import tqdm

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Reproducibility
seed = 42
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)
torch.cuda.manual_seed_all(seed)

# =========================
# Paths & Hyperparams
# =========================
# Set this to your Malimg root where subfolders are class names:
root_dir = "/content/drive/MyDrive/Colab Notebooks/malimg_paper_dataset_imgs"

batch_size = 128
pretrain_epochs = 100
finetune_epochs = 100     # early stopping can cut this short
learning_rate = 1e-3     # for SimCLR
finetune_lr  = 1e-4      # for classifier + encoder
temperature  = 0.5
projection_dim = 128
test_ratio = 0.2
patience = 5             # early stopping on train loss (no val per your request)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# =========================
# Dataset helpers
# =========================
def list_images_and_labels(root_dir):
    class_names = sorted([d for d in os.listdir(root_dir)
                          if os.path.isdir(os.path.join(root_dir, d))])
    class_to_idx = {c:i for i,c in enumerate(class_names)}
    img_paths, labels = [], []
    for cls in class_names:
        cls_path = os.path.join(root_dir, cls)
        for f in os.listdir(cls_path):
            p = os.path.join(cls_path, f)
            if os.path.isfile(p):
                img_paths.append(p)
                labels.append(class_to_idx[cls])
    return img_paths, labels, class_names

class MalimgDatasetFromList(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img = Image.open(self.image_paths[idx]).convert("RGB")
        label = self.labels[idx]
        if self.transform:
            img = self.transform(img)
        return img, label

class ContrastiveDataset(Dataset):
    """ Wrap a base dataset & return two augmented views for SimCLR. """
    def __init__(self, base_dataset, augment):
        self.base = base_dataset
        self.augment = augment

    def __len__(self):
        return len(self.base)

    def __getitem__(self, idx):
        img, _ = self.base[idx]     # base returns (PIL-transformed tensor, label)
        # If base returns tensors already, ensure augment expects PIL -> ToTensor inside
        # Here, we want PIL input; so build base without transform, and put transforms here.
        # To make that work, ensure base has transform=None (we will below).
        # We'll handle it by reloading image path via base.image_paths if needed.
        # Simpler: create a PIL loader dataset for contrastive step:
        # We'll reconstruct PIL image by path to guarantee augment works.

        # Re-open PIL to ensure augment gets PIL (in case base had ToTensor)
        # (This requires base to have .image_paths; we can check and fallback.)
        if hasattr(self.base, "image_paths"):
            pil = Image.open(self.base.image_paths[idx]).convert("RGB")
        else:
            # Fallback: assume img is PIL already
            pil = img

        x1 = self.augment(pil)
        x2 = self.augment(pil)
        return x1, x2

# =========================
# Model (SimCLR wrapper)
# =========================
class SimCLR(nn.Module):
    def __init__(self, base_model, projection_dim=128):
        super().__init__()
        self.encoder = base_model  # resnet50 outputs 1000-dim
        self.projection = nn.Sequential(
            nn.Linear(1000, 512),
            nn.ReLU(inplace=True),
            nn.Linear(512, projection_dim)
        )

    def forward(self, x):
        h = self.encoder(x)    # (N, 1000)
        z = self.projection(h) # (N, projection_dim)
        return h, z

def nt_xent_loss(z1, z2, temperature=0.5):
    N = z1.size(0)
    z1 = F.normalize(z1, dim=1)
    z2 = F.normalize(z2, dim=1)
    z  = torch.cat([z1, z2], dim=0)   # (2N, D)

    sim = torch.matmul(z, z.T) / temperature         # (2N, 2N)
    mask = torch.eye(2*N, dtype=torch.bool, device=z.device)
    sim = sim.masked_fill(mask, -9e15)               # remove self-similarity

    # positives: for every i in [0..N-1], pair i with i+N and vice versa
    pos = torch.cat([torch.diag(sim, N), torch.diag(sim, -N)], dim=0)  # (2N,)
    # negatives: the rest
    neg = sim

    logits = torch.cat([pos.unsqueeze(1), neg], dim=1)  # (2N, 1 + (2N-1))
    labels = torch.zeros(2*N, dtype=torch.long, device=z.device)  # positives at index 0
    loss = F.cross_entropy(logits, labels)
    return loss

# =========================
# Data prep + split
# =========================
all_paths, all_labels, class_names = list_images_and_labels(root_dir)
num_classes = len(class_names)

train_paths, test_paths, train_labels, test_labels = train_test_split(
    all_paths, all_labels,
    test_size=test_ratio,
    stratify=all_labels,
    random_state=seed
)

# For SimCLR pretraining we need PIL images (no tensor yet)
train_dataset_pil = MalimgDatasetFromList(train_paths, train_labels, transform=None)

# Augmentations for SimCLR
simclr_augment = T.Compose([
    T.RandomResizedCrop(224),
    T.RandomHorizontalFlip(),
    T.RandomApply([T.ColorJitter(0.4, 0.4, 0.4, 0.1)], p=0.8),
    T.RandomGrayscale(p=0.2),
    T.ToTensor()
])

contrastive_dataset = ContrastiveDataset(train_dataset_pil, simclr_augment)

# WeightedRandomSampler (train split)
class_counts = np.bincount(train_labels, minlength=num_classes).astype(np.float32)
class_counts[class_counts == 0] = 1.0
weights_per_class = 1.0 / class_counts
sample_weights = np.array([weights_per_class[y] for y in train_labels], dtype=np.float32)
train_sampler = WeightedRandomSampler(
    weights=torch.as_tensor(sample_weights),
    num_samples=len(sample_weights),
    replacement=True
)

contrastive_loader = DataLoader(
    contrastive_dataset,
    batch_size=batch_size,
    sampler=train_sampler,
    num_workers=2,
    pin_memory=True
)

# Classification datasets/loaders
train_transform_cls = T.Compose([
    T.Resize((224, 224)),
    T.ToTensor()
])
test_transform_cls = T.Compose([
    T.Resize((224, 224)),
    T.ToTensor()
])

train_dataset_cls = MalimgDatasetFromList(train_paths, train_labels, transform=train_transform_cls)
test_dataset_cls  = MalimgDatasetFromList(test_paths,  test_labels,  transform=test_transform_cls)

# Use the same sampler for classification training loader
train_loader_cls = DataLoader(
    train_dataset_cls,
    batch_size=batch_size,
    sampler=train_sampler,
    num_workers=2,
    pin_memory=True
)
test_loader = DataLoader(
    test_dataset_cls,
    batch_size=batch_size,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)

# =========================
# Build models
# =========================
base_model = models.resnet50(pretrained=True)  # keep same
simclr_model = SimCLR(base_model, projection_dim=projection_dim).to(device)

optimizer = optim.Adam(simclr_model.parameters(), lr=learning_rate)

# =========================
# Pretraining (SimCLR) + timing
# =========================
pretrain_start = time.time()
best_train_loss = float("inf")
no_improve = 0

for epoch in range(pretrain_epochs):
    simclr_model.train()
    running_loss = 0.0
    for (x1, x2) in tqdm(contrastive_loader, desc=f"SimCLR Pretrain {epoch+1}/{pretrain_epochs}"):
        x1, x2 = x1.to(device, non_blocking=True), x2.to(device, non_blocking=True)

        _, z1 = simclr_model(x1)
        _, z2 = simclr_model(x2)
        loss = nt_xent_loss(z1, z2, temperature=temperature)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    epoch_loss = running_loss / max(1, len(contrastive_loader))
    print(f"[Pretrain] Epoch {epoch+1}/{pretrain_epochs} | Loss: {epoch_loss:.4f}")

    # (Optional) Early stop on SimCLR training loss as well
    if epoch_loss + 1e-6 < best_train_loss:
        best_train_loss = epoch_loss
        no_improve = 0
    else:
        no_improve += 1
        if no_improve >= patience:
            print(f"Early stopping SimCLR pretraining at epoch {epoch+1} (no improvement for {patience} epochs).")
            break

if device.type == "cuda":
    torch.cuda.synchronize()
pretrain_time = time.time() - pretrain_start
print(f"Pretraining time: {pretrain_time/60:.2f} min")

# =========================
# Fine-tune classifier (train split only) with early stopping + timing
# =========================
classifier = nn.Linear(1000, num_classes).to(device)
criterion  = nn.CrossEntropyLoss()
optimizer_cls = optim.Adam(
    list(simclr_model.encoder.parameters()) + list(classifier.parameters()),
    lr=finetune_lr
)

best_loss = float("inf")
no_improve = 0

ft_start = time.time()
for epoch in range(finetune_epochs):
    simclr_model.train()
    classifier.train()
    running_loss = 0.0
    correct, total = 0, 0

    for imgs, labels in tqdm(train_loader_cls, desc=f"Fine-tune {epoch+1}/{finetune_epochs}"):
        imgs = imgs.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)

        h, _ = simclr_model(imgs)
        outputs = classifier(h)
        loss = criterion(outputs, labels)

        optimizer_cls.zero_grad()
        loss.backward()
        optimizer_cls.step()

        running_loss += loss.item()

        # Train accuracy
        _, preds = torch.max(outputs, 1)
        correct += (preds == labels).sum().item()
        total   += labels.size(0)

    epoch_loss = running_loss / max(1, len(train_loader_cls))
    train_acc  = 100.0 * correct / max(1, total)
    print(f"[Fine-tune] Epoch {epoch+1}/{finetune_epochs} | Loss: {epoch_loss:.4f} | Train Acc: {train_acc:.2f}%")

    # Early stopping on training loss (since no val set requested)
    if epoch_loss + 1e-6 < best_loss:
        best_loss = epoch_loss
        no_improve = 0
        # Save best weights
        torch.save(simclr_model.encoder.state_dict(), "/content/drive/MyDrive/Colab Notebooks/SimCLR/simclr_resnet50_encoder_best.pth")
        torch.save(classifier.state_dict(), "/content/drive/MyDrive/Colab Notebooks/SimCLR/simclr_resnet50_classifier_best.pth")
    else:
        no_improve += 1
        if no_improve >= patience:
            print(f"Early stopping fine-tuning at epoch {epoch+1} (no improvement for {patience} epochs).")
            break

if device.type == "cuda":
    torch.cuda.synchronize()
finetune_time = time.time() - ft_start
print(f"Fine-tune time: {finetune_time/60:.2f} min")
print(f"Total training time: {(pretrain_time + finetune_time)/60:.2f} min")

# Load best weights (from fine-tune)
simclr_model.encoder.load_state_dict(torch.load("/content/drive/MyDrive/Colab Notebooks/SimCLR/simclr_resnet50_encoder_best.pth"))
classifier.load_state_dict(torch.load("/content/drive/MyDrive/Colab Notebooks/SimCLR/simclr_resnet50_classifier_best.pth"))
simclr_model.eval()
classifier.eval()

# =========================
# Test Evaluation (accuracy, report, confusion matrix, avg inference time)
# =========================
all_preds, all_labels = [], []
inference_images = 0
total_infer_time = 0.0

with torch.no_grad():
    for imgs, labels in tqdm(test_loader, desc="Testing"):
        imgs = imgs.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)

        if device.type == "cuda":
            torch.cuda.synchronize()
        t0 = time.time()

        h, _ = simclr_model(imgs)
        outputs = classifier(h)

        if device.type == "cuda":
            torch.cuda.synchronize()
        t1 = time.time()

        total_infer_time += (t1 - t0)
        inference_images += imgs.size(0)

        _, preds = torch.max(outputs, 1)
        all_preds.extend(preds.cpu().numpy().tolist())
        all_labels.extend(labels.cpu().numpy().tolist())

test_acc = accuracy_score(all_labels, all_preds) * 100.0
avg_infer_ms = (total_infer_time / max(1, inference_images)) * 1000.0
print(f"Test Accuracy: {test_acc:.2f}%")
print(f"Average Inference Time (per image): {avg_infer_ms:.3f} ms")

print("Classification Report:")
print(classification_report(all_labels, all_preds, target_names=class_names, digits=4))

# Confusion Matrix (matplotlib only, no seaborn)
cm = confusion_matrix(all_labels, all_preds, labels=list(range(num_classes)))

plt.figure(figsize=(10, 8))
plt.imshow(cm, interpolation='nearest')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.colorbar()
tick_marks = np.arange(num_classes)
plt.xticks(tick_marks, class_names, rotation=90)
plt.yticks(tick_marks, class_names)
plt.tight_layout()
plt.show()

# =========================
# t-SNE on test features (from encoder output h)
# =========================
features, labels_tsne = [], []
with torch.no_grad():
    for imgs, labels in tqdm(test_loader, desc="Extracting features for t-SNE"):
        imgs = imgs.to(device, non_blocking=True)
        h, _ = simclr_model(imgs)        # h is (N, 1000)
        features.append(h.cpu().numpy())
        labels_tsne.extend(labels.numpy())
features = np.concatenate(features, axis=0)
labels_tsne = np.array(labels_tsne)

tsne = TSNE(n_components=2, perplexity=30, learning_rate=200, random_state=seed, n_iter=1000)
features_2d = tsne.fit_transform(features)

plt.figure(figsize=(10, 8))
scatter = plt.scatter(features_2d[:, 0], features_2d[:, 1], c=labels_tsne, s=10, alpha=0.7)
plt.title("t-SNE of Encoder Features (Test Set)")
plt.tight_layout()
plt.show()

# =========================
# Final save
# =========================
torch.save(simclr_model.encoder.state_dict(), "/content/drive/MyDrive/Colab Notebooks/SimCLRsimclr_resnet50_encoder_final.pth")
torch.save(classifier.state_dict(),          "/content/drive/MyDrive/Colab Notebooks/SimCLR/simclr_resnet50_classifier_final.pth")
print("Done. Saved best & final weights.")
```
